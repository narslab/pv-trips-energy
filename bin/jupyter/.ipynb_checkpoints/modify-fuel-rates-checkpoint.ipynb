```python
import pandas as pd
```

```python
df_final = pd.read_csv(r'../../results/fuel-rates-CDB-CHB.csv', delimiter=',', skiprows=0, low_memory=False)
```

```python
df_final.columns
```

```python
df_final.Vehicle_mass
```

```python
zeros = df_final[df_final['FuelRate(L/s)'] == 0]
zeros
```

```python
for i in df_final.index:
    if i==0:
        df_final.at[i , 'FuelRate(L/s)'] = 0
    elif df_final['Vehicle_mass']==0:
        df_final.at[i , 'FuelRate(L/s)'] = 0
    else:
        if df_final['Vehicle'].loc[i]==df_final['Vehicle'].loc[i-1]:
            if df_final['Date'].loc[i]!=df_final['Date'].loc[i-1]:
                df_final.at[i , 'FuelRate(L/s)'] = 0
        else:
            df_final.at[i , 'FuelRate(L/s)'] = 0  
```
