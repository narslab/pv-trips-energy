```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
df1 = pd.read_csv(r'../../results/fuel-rates-CDB-CHB.csv', delimiter=',', skiprows=0, low_memory=False)
```

```python
df1
```

```python
df1.loc[df1['FuelRate(L/s)']>=10]
```

```python
figure(figsize=(14, 9.5), dpi=400)
plt.plot_date(x=df1['ServiceDateTime']  , y=df1['FuelRate(L/s)'], fmt='o', tz=None, xdate=True, ydate=False, color='teal')
plt.ylabel("FuelRate(L/s)", fontsize=36)
plt.xticks(fontsize=28, rotation = 45)
plt.yticks(fontsize=28)
plt.savefig(r'../../figures/FuelRate-CDB-CHB.png', dpi=400, bbox_inches='tight')
plt.show()
```

```python
df_filtered=df1.loc[df1['FuelRate(L/s)']<=10]
```

```python
figure(figsize=(14, 9.5), dpi=400)
plt.plot_date(x=df_filtered['ServiceDateTime']  , y=df_filtered['FuelRate(L/s)'], fmt='o', tz=None, xdate=True, ydate=False, color='teal')
plt.ylabel("FuelRate(L/s)", fontsize=36)
plt.xticks(fontsize=28, rotation = 45)
plt.yticks(fontsize=28)
plt.savefig(r'../../figures/FuelRate-filtered-CDB-CHB.png', dpi=400, bbox_inches='tight')
plt.show()
```
