```python
import pandas as pd
```

```python
df_final = pd.read_csv(r'../../results/computed-fuel-rates.csv', delimiter=',', skiprows=0, low_memory=False)
```

```python
df_final.sort_values(by=['Vehicle','ServiceDateTime'], ascending=True, inplace=True)
```

```python
df_final
```

```python
for i in df_final.index:
    if i==0:
        df_final.at[i , 'FuelRate/Energy'] = 0
    elif df_final['Vehicle_mass'].loc[i]==0:
        df_final.at[i , 'FuelRate/Energy'] = 0
    else:
        if df_final['Vehicle'].loc[i]==df_final['Vehicle'].loc[i-1]:
            if df_final['Date'].loc[i]!=df_final['Date'].loc[i-1]:
                df_final.at[i , 'FuelRate/Energy'] = 0
        else:
            df_final.at[i , 'FuelRate/Energy'] = 0  
```
